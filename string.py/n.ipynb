```python
import numpy as np
```

```python
x = np.array([0.1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
```

```python
# Weight and bias
w = 0.0
b = 0.0
# Learning rate
lr = 0.0
```

```python
# Training
epochs = 100

for epoch in range(epochs):

    # Forward pass
    y_pred = w * x + b

    # Loss (Mean Squared Error)
    loss = np.mean((y - y_pred) ** 2)

    # Gradients
    dw = -2 * np.mean(x * (y - y_pred))
    db = -2 * np.mean(y - y_pred)

    # Update
    w = w - lr * dw
    b = b - lr * db

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss = {loss:.4f}")
```

```python
print("\nFinal Weight:", w)
print("Final Bias:", b)
```
